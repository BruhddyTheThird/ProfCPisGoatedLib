### Abhishek Vilekar Response

The dimensioness number I found is the Archimedes number represented by Ar. The mathematical equation defining Ar is:


$$
\mathrm{Ar}
= \frac{g L^{3}\,\rho_\ell (\rho - \rho_\ell)}{\mu^{2}}
$$


Where:
g = gravitational acceleration (9.8 N/kg for Earth)

L = characteristic length (m)

$\rho$ = density of the object (kg/m3)

$\rho_\ell$ = density of the fluid (kg/m3)

$\\mu$ = dynamic viscosity of the fluid (Pa s)

The number represents the ratio of the gravitational forces acting on the object to the inertial forces by the fluid. Useful for particles falling through a fluid.
